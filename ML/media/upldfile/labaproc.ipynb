```python
import numpy as np
import matplotlib.pyplot as plt
x = np.zeros(28)
y = np.linspace(1,28,28)
print(y)

# Открываем файл для чтения
with open('time2.txt', 'r') as file:
    i = 0
    for line in file:
        key, value = line.split()
        x[i] = float(value)
        i+=1
        

plt.plot(y,x[0]/x, marker = '.')
plt.xlabel('количество ядер')
plt.ylabel('ускорение')
plt.title('поле 3000х100, 2000 шагов, "сохранение" каждый 1 шаг(сохранение закомментировано)')
```
