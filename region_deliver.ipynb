```python
clients = pd.read_excel('clients.xlsx')
clients.head(2)
```

```python
# Строки с двумя адресами. Их всего 4, так что уберём их

print(len(clients[clients['Адрес грузополучателя'].str.contains(';') == True]))
clients.drop(clients[clients['Адрес грузополучателя'].str.contains(';') == True].index, inplace=True)
```

```python
clients[[f'{i}' for i in range(11)]] = clients['Адрес грузополучателя'].str.split(',', expand=True)
clients.drop('Адрес грузополучателя', axis=1, inplace=True)
clients.fillna('-', inplace=True)
clients.head(2)
```

```python
cols = clients.columns[1:12]
```

```python
# Поиск индекса, города, улицы, дома
import re


def cond(row):
    if row == []:
        return None
    else:
        return row[0]


clients['index'] = [ [clients.iloc[j, i].strip()
for i in range(1, len(cols))
if bool(re.match(r'\d{6}$', clients.iloc[j, i].strip())) ]
                     for j in range(len(clients))]
clients['index'] = clients['index'].apply(cond)

clients['city'] = [ [clients.iloc[j, i].strip()
for i in range(1, len(cols))
if bool(re.match(r'г.\s*\w+[-]*\w*$', clients.iloc[j, i].strip())) ]
                     for j in range(len(clients))]
clients['city'] = clients['city'].apply(cond)

clients['street'] = [ [clients.iloc[j, i].strip()
for i in range(1, len(cols))
if bool(re.match(r'ул.\s\w+|пр.\s\w+|\w+\sбульвар|\w+\sпер.', clients.iloc[j, i].strip())) ]
                     for j in range(len(clients))]
clients['street'] = clients['street'].apply(cond)

clients['house'] = [ [clients.iloc[j, i].strip()
for i in range(1, len(cols))
if bool(re.match(r'д.\s*\w+[.]*$', clients.iloc[j, i].strip())) ]
                     for j in range(len(clients))]
clients['house'] = clients['house'].apply(cond)

clients.head(20)
```

```python
clients_d = clients.drop([str(i) for i in range(11)], axis=1).fillna('-').drop(clients[(clients['city'].fillna('-') == '-') & (clients['index'].fillna('-') == '-')].index)
str( round(1 - len(clients_d) / len(clients), 5) ) + ': Доля грузополучателей без обнаруженных города и индекса'
```

```python
same_clients = set(df2_d['Грузополучатель']) & set(clients['Код грузополучателя'])
str(len(same_clients)) + ': количество грузополучателей, имеющихся в таблице с поставками и в справочнике грузополучателей'
```

```python
str(round(len(same_clients) / len(clients['Код грузополучателя'].unique()), 4)) + ': доля уникальных грузополучателей, имеющихся в таблице с поставками и в справочнике грузополучателей'
```

```python
import pprint

df2_d = df2_d.rename(columns={'Грузополучатель': 'Код грузополучателя'})
df2_dm = df2_d.merge(clients_d, on='Код грузополучателя')
df2_dm.drop(df2_dm[df2_dm['Поставщик'].isna() == True].index, inplace=True) # ['Код грузополучателя'].unique().size  # = 109

pprint.pp(str(df2_dm['Код грузополучателя'].unique().size / len(same_clients)) + ': Доля грузополучателей, имеющихся в таблице с поставками и в справочнике грузополучателей, у которых обнаружены хотя бы город или индекс, а также имеются данные о поставщике')
```
